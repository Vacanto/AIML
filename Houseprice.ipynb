```python
from sklearn.model_selection import train_test_split
```

```python
df_encoded = pd.get_dummies(data, drop_first=True)
df_f = df_encoded.fillna(0)
```

```python
df_f
```

```python
X=df_f.drop("SalePrice",axis=1)
Y=df_f["SalePrice"]
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.40,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
la= LinearRegression()
```

```python
la.fit(X_train,Y_train)
```

```python
coefficients = la.coef_
```

```python
coeff_df = pd.DataFrame(coefficients, index=X_train.columns, columns=['Coeff'])
```

```python
coeff_df
```

```python
prediction =la.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,prediction)
```
